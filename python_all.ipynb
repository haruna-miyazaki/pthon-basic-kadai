```python

class Human:
  def __init__(self, name, age):
    self.name =name
    self.age =age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name} is an adult.")
    else:
      print(f"{self.name} is not an adult.")


humans = [ Human("arice", 25), Human("bob", 18)]

for human in humans:
  human.check_adult()
```
